## Переменные и их обзывательства

При названии переменных нужно опираться на ряд правил (в большинстве своем они интуетивные, запомните быстро):
1) Имя переменной должно начинаться с буквы (`peremennaya`) или символа подчеркивания (`_peremennaya`)
2) Имя переменной не может начинаться с цифры
3) Имя переменной не может содержать специальные символы потипу `!?*%#@`
4) Разные регистры разные переменные: (`Peremennaya`), (`PEREMENNAYA`), (`PeremennayA`), (`PERemennaya`), (`peremennaya`)

- Это правило не строгое, но настоятельно рекомендую его придерживаться: **Называйте переменные так, чтобы описать что она будет в себе хранить**

*Например:* `phrase = 'Привет!' - хранит в себе какую-то фразу`, `mult = 4 - хранит в себе какой-то множитель`

## Операции

### Над числами

```python
summ = 2 + 2 # Сложение
print('2+2 =',summ)

diff = 2 - 2 # Вычетание
print('2-2 =',diff)

mult = 2 * 2 # Умножение
print('2*2 =',mult)

division = 2 / 2 # Деление - результат всегда float
print('2/2 =',division)

pow = 2 ** 2 # Возведение в степень
print('2**2 =', pow)

mod = 7 % 3 # Остаток от деления
print('7 % 3 =', mod)

div = 7 // 3 # Деление нацело
print('7 // 3 =', div)
```

**Упрощенные операции над перемеными**

```python
x = 5
print('x =', x)

x += 2 # x = x + 2
print('x += 2 -> x =', x)
x -= 2 # x = x - 2
print('x -= 2 -> x =', x)   
x *= 2 # x = x * 2
print('x *= 2 -> x =', x)
x /= 2 # x = x / 2
print('x /= 2 -> x =', x)
```

**Логические операторы**

```python
a = 5
b = 10

print('a больше b?', a > b) # >=
print('a меньше b?', a < b) # <=
print('a равно b?', a == b) 

# Функция not() отрицает логическое значение, возвращая True, если значение ложно, и False, если значение истинно
print('a не равно b?', not a == b) # !=
# или
print('a не равно b?', a != b)

print('a больше b?', a > b and a != b) # логическое И - возвращает True, если оба условия истинны &
print('a больше b?', a > b or a != b) # логическое ИЛИ - возвращает True, если хотя бы одно из условий истинно |
```

### **for**
![for](материалы/for.png)

Цикл *for* используется, когда количество итераций заранее известно. Он выполняет код для каждого элемента заданной последовательности.

Обычно в цикле *for* используется итерируемая переменная. Чаще всего её называют *i*, но можно использовать любое осмысленное имя. На каждой итерации переменная принимает новое значение из диапазона.

Чтобы задать диапазон значений, в Python используется функция `range()`.

Строго говоря, `range()` — это не функция, а встроенный тип данных (Built-in Type), который создаёт последовательность целых чисел. Подробнее про него можно почитать в документации.

Но для простоты мы будем называть его функцией.

```python
spisok = ['Первый элемент', 'Второй элемент', 'Третий элемент']

for element in spisok: # оператор for используется для перебора элементов в коллекции (списке, строке, словаре и т.д.)
    print(element) # в каждой итерации цикла переменная element принимает значение следующего элемента из списка spisok, и мы выводим его на экран
```

```python
for i in range(5):
    print(i)
```

```python
message = "Привет, мир!"
for char in message:
    print(char)
```

```python
spisok = [1, 2, 3, 4, 5]

for number in spisok: # Переменная number может называтся как угодно, обычно используют i, j, k для обозначения переменных цикла, но это не обязательно
    print(number) # в каждой итерации цикла переменная number принимает значение следующего элемента из списка spisok, и мы выводим его на экран
```

Для того чтобы не создавать список заранее, используем `range()`

```python
for i in range(5): # функция range() генерирует последовательность чисел от 0 до 4 (не включая 5)
    print(i) # в каждой итерации цикла переменная i принимает значение следующего числа из последовательности, и мы выводим его на экран
```

```python
for i in range(1, 6, 2):  # функция range() принимает 3 аргумента: начальное значение, конечное значение и шаг. Если указать только одно число, то начальное значение будет 0, а шаг будет 1. 
    print(i)
```

```python
n = int(input("Введите первое число: "))
for i in range(n, -1, -1): # функция range() генерирует последовательность чисел от n до -1 (не включая -1) с шагом -1, то есть в обратном порядке
    print(i)
```

#### **Практика**

![practice](материалы/practice.png)

У вас есть список оценок ученика Богдана: `grades = [5, 3, 4, 2, 5, 4, 3, 5, 2, 4]`
вам необходимо вывести:
1) Сумму всех оценок
2) Средний балл
3) сколько троек получил Богдан

```python
grades = [5, 3, 4, 2, 5, 4, 3, 5, 2, 4]

suma = 0
col_grades = 0
col_three = 0

for grade in grades:
    suma = suma + grade
    col_grades = col_grades + 1
    
    if grade == 3:
        col_three += 1 # col_three = col_three + 1

print(suma)
print(suma / col_grades)
print(col_three)

```

```python
for i, grade in enumerate(grades): # функция enumerate() возвращает пару индекс-значение для каждого элемента в списке grades, и мы распаковываем эту пару в переменные i и grade
    print(f'Оценка номер {i} равна {grade}') # в каждой итерации цикла переменная i принимает значение индекса следующей оценки из списка grades, а переменная grade принимает значение самой оценки, и мы выводим их на экран
```

#### **Практика**

![practice](материалы/practice.png)

Нужно написать функцию для установки пароля из цифр, чтобы пользователь вводил значение и оно проверялось внутри функции
на надежность, а именно:
1) Пароль должен быть не короче 8 символов
2) Пароль должен содержать только цифры (используйте переменную `valid_chars = "0123456789"`)

Если пароль удовлетворяет этим требованиям, в таком случае выведите "пароль сохранен" 

```python
def pass_check(password):
    valid_chars = "0123456789"

    if len(password) < 8: 
        return "Пароль не подходит, нужно 8 или более символов"
    
    for symb in password:
        if symb not in valid_chars:
            return "Пароль не подходит, только цифры"

    return "Пароль сохранен"
    

print(pass_check(input("Введите пароль:")))

```

## Библиотеки

```python
# для подключения библиотек используется оператор import
import random # импортируем библиотеку random для работы с генерацией случайных чисел


print(random.randint(1, 10)) # пример использования функции randint() из библиотеки random для генерации случайного числа от 1 до 10
```

```python
# для удобства работы с библиотекой, мы может изменить название библиотеки, используя оператор as
import random as rd # теперь мы можем использовать сокращенное название rd для обращения к функциям библиотеки random

print(rd.randint(1, 10)) # пример использования функции randint() из библиотеки random для генерации случайного числа от 1 до 10
```

```python
# Если нам нужно использовать только одну функцию из библиотеки, то мы можем импортировать только эту функцию, используя синтаксис: from название_библиотеки import название_функции
from random import randint # теперь мы можем использовать функцию randint() без указания названия библиотеки

print(randint(1, 1000)) # пример использования функции randint() для генерации случайного числа от 1 до 10
```
