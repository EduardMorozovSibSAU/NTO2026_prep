```python
n, m = map(int, input().split())
x, y = map(int, input().split())

def count_ways(n, m):
    ways = [[0 for i in range(m)] for i in range(n)]

    for i in range(n): # итерируемся по строкам
        for j in range(m): # итерируемся по столбцам         [0,0,0,0]
            ways[i][j] = j + i
    return ways[i][j]

ways_to_xy = count_ways(x, y)
ways_to_end = count_ways(n - x + 1, m - y + 1)

print(ways_to_xy * ways_to_end)
```
