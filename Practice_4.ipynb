```python
n = int(input())
lengths = map(int ,input().split())

# n = 3 
# lengths = [4, 8, 2, 3, 6, 7] # 2n 

lengths_sorted = sorted(lengths)
print(lengths_sorted[0] + lengths_sorted[-1])


pairs = []
for i in range(len(lengths) // 2): 

    val_start = lengths_sorted[i]
    val_end = lengths_sorted[-i-1]

    indx_first = lengths.index(val_start) + 1
    indx_second = lengths.index(val_end) + 1

    pairs.append([indx_first, indx_second])

pairs = sorted(pairs)

for i in range(n): 
    print(pairs[i][0], pairs[i][1])
    
```

```python
# NTO 2

n = 3
lengths = [4, 8, 2, 3, 6, 7]

sorted_lengths = sorted(lengths)
print(sorted_lengths[0] + sorted_lengths[-1])

# print(f'sorted_lengths: {sorted_lengths}')

pairs = []
for i in range(len(sorted_lengths)//2):

    value1 = sorted_lengths[i]
    value2 = sorted_lengths[-i-1]

    index_first_piece = lengths.index(value1)+1
    index_second_piece = lengths.index(value2)+1
    
    pair = [index_first_piece, index_second_piece]
    pairs.append(pair)
    
pairs.sort()
for pair in pairs:
    print(pair[0], pair[1])
```

```python
# NTO 3
a, b, c, d = 4, 10, 9, 6

min_value = min(a, b, c, d) # Минимальный размер квадратов
max_from_glass1 = min(max(a, b) / 2, min(a,b)) # максимальный размер квадратов которые можно вырезать из первого прямоугольника
max_from_glass2 = min(max(c, d) / 2, min(c,d)) # максимальный размер квадратов которые можно вырезать из второго прямоугольника

print(max(max_from_glass1, max_from_glass2, min_value))
```

```python
# a, b, c, d = map(int, input().split())
# n, m = map(int, input().split())
a, b, c, d = 8, 20, 4, 2
n, m = 3, 4



full_cells = min(a, c) +  min(b, d)
rest_tri = abs(a - c) + abs(b - d)

# print(rest_tri)
count_cells = n * m

if full_cells >= count_cells:
    print(full_cells * 2)
else:
    half_empty_cells = count_cells - full_cells
    
    if rest_tri < half_empty_cells:
        print(full_cells * 2 + rest_tri)
    else:
        print(full_cells * 2 + half_empty_cells)


```
