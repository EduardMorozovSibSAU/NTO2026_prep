### O(n) Линейная сложность


```python
def findMax(arr): # функция для нахождения максимального элемента в массиве
    max = arr[0]
    for i in range(1, len(arr)):
        if (arr[i] > max):
            max = arr[i]
    return max
```

### Логарифмическая сложность O(log n)

```python
def binarySearch(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid  # элемент найден
        elif arr[mid] < target:
            left = mid + 1  # искать в правой половине  
        else:
            right = mid - 1  # искать в левой половине  
    return -1  # элемент не найден
```

```python
def swap(arr, a, b): # функция смены
    temp = arr[a]
    arr[a] = arr[b]
    arr[b] = temp
```

### Сортировка выбором
Временная сложность = O(n²).
Внутренний цикл работает не менее n раз. Поэтому операция займет не менее n² времени

Пространственная сложность = O(1).
Дополнительная память не используется, потому что происходит обмен элементов в исходном массиве

```python

mas2sort = [3, 1, 7, 4, 6, 8, 2, 5]
n = len(mas2sort)
    
# итерируемся по массиву
for i in range(0, n):
    current_min = mas2sort[i]  # инициализируемся первым значеним 
    min_index = i  # инициализируем минимальный индекс
    
    for j in range(i, n):  # проходимся по оставшимся элементам массива
        
        if mas2sort[j] < current_min:  # проверяем, если j-тое значение меньше текушего минимального    
            
            current_min = mas2sort[j]  # обновляес минимальные значение и индекс
            min_index = j
            
    # меняем i-тое и j-тое значения
    swap(mas2sort, i, min_index)

print(mas2sort)
```

### Сортировка пузырьком
Временная сложность = O(n²).
n² итераций очевидны из двух вложенных циклов.

Пространственная сложность = O(1).
Как упоминалось выше, сортировка происходит в том же массиве, поэтому использование памяти не зависит от данных обработки.

```python
mas2sort = [3, 1, 7, 4, 6, 8, 2, 5]
n = len(mas2sort)

# итерируемся по неотсорт. массиву до предпоследнего элемента
for i in range(n):
    
    # проставляем условия флага для финального списка
    swapped = False
    
    # итерируемся по осташвимся неотсортированным объектам
    for j in range(n - 1 - i):
        
        # сравниваем соседние элементы
        if mas2sort[j] > mas2sort[j + 1]: # т.к. мы смотрим вперед, нужно брать j-1 чтобы не выходить за размер массива
            
            # меняем элементы местами
            swap(mas2sort, j, j + 1)
            swapped = True
    
    # завершаем алгоритм, если смены не произошло
    if not swapped:
        break

print(mas2sort)    
```

## Быстрая сортировка

```python
def quick_sort(unsorted, start, end):
    """ быстрая сортировка """

    # останавливаемся, когда индекс слева достиг или превысил индек справа
    if start >= end:
        return 

    # определяем позицию следующей оси
    i_pivot = partition(unsorted, start, end - 1)

    # рекурсивный вызов левой части
    quick_sort(unsorted, start, i_pivot)

    # рекурсивный вызов правой части
    quick_sort(unsorted, i_pivot + 1, end)

    return unsorted

def partition(unsorted, start, end):
    """ arrange (Левый массив < pivot) and (Левый массив > pivot) """

    # выбираем значение pivot как последний элемент неотсортированного сегмента
    pivot = unsorted[end]

    # назначаем на pivot значение левого индекса
    i_pivot = start

    # проходим от начала до конца текущего сегмента
    for i in range(start, end):

        # сравниваем текущее значение со значением pivot
        if unsorted[i] <= pivot:

            # меняем местами текущее значение и значенрие pivot
            swap(unsorted, i, i_pivot)

            # увеличиваем значение пивота
            i_pivot += 1

    # ставим пивот в правильную позицию, заменив со значением слева
    swap(unsorted, i_pivot, end)

    # возвращаем следующее значение пивота
    return i_pivot

quick_sort(unsorted = mas2sort, start = 0, end = len(mas2sort))
```
